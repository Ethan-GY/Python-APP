```python
#our question is :The relationship between students' weekly attendence and their final exam scores.
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("StudentPerformanceFactors.csv")
df=df.groupby('Attendance')['Exam_Score'].mean().reset_index()
plt.plot(df['Attendance'],df['Exam_Score'],color='red')
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.title('Relationship between Attendance and Exam Score')
plt.show()
```

```python
###Relationship between sleep hours and exam scores
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("StudentPerformanceFactors.csv")
df=df.groupby('Sleep_Hours')['Exam_Score'].mean().reset_index()
plt.plot(df['Sleep_Hours'],df['Exam_Score'],color='red')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.title('Relationship between Sleep Hours and Exam Score')
plt.show()
```

```python
#The distribution of sleep hours among students
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("StudentPerformanceFactors.csv")
df=df.groupby('Sleep_Hours').count().reset_index()
plt.pie(df['Sleep_Hours'], labels=df['Sleep_Hours'], autopct='%1.1f%%')
plt.title('Distribution of Sleep Hours Among Students')
plt.show()
```
